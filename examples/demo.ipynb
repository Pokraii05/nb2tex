# Theoretical overview

A real voltage source can be modeled as an ideal electromotive force (emf) $U_q$ in series with an internal resistance $R_i$. The terminal voltage $U_k$ across the source depends on the current $I$ drawn by the load resistor $R_L$.

$$U_k = U_q - I \, R_i$$

The short-circuit current $I_k$ (when $U_k = 0$) is therefore:

$$I_k = \frac{U_q}{R_i}$$

When no current flows ($I = 0$), the terminal voltage equals the emf:

$$U_k = U_q$$

This linear relationship between $U_k$ and $I$ allows one to determine both $R_i$ and $U_q$ by fitting a straight line to the measured voltage-current data.
